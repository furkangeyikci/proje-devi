```python
g=int(input("PO4-3 miktarını giriniz"))
h=int(input("Çözelti hacmini giriniz"))
Fosfat_Miktarı =(g/h)
print("Fosfat değeri", Fosfat_Miktarı ,"mg P/L")

f=int(input("Amonyum nitrojeni miktarını giriniz"))
e=int(input("Çözelti hacmini giriniz"))
Amonyum_Azotu =(f/e)
print("Amonyum Azotu değeri", Amonyum_Azotu ,"mg N/L")

d=int(input("suya giren oksijen miktarını mg cinsinden giriniz"))
c=int(input("5 gün sonraki sudan çıkan oksijen miktarını mg cinsinden giriniz"))
BOİ =(d-c)
print("BOİ değeri" ,BOİ ,"mg/L")

a=int(input("suya giren oksijen miktarını mg cinsinden giriniz"))
b=int(input("sudan çıkan oksijen miktarını mg cinsinden giriniz"))
KOİ =(a-b)
print("koi değeri", KOİ ,"mg/L")

import math

def hesapla_pH(m):
    pH = math.log10(m)
    return pH

m = float(input("Hidrojen iyonu konsantrasyonunu giriniz: "))

sonuc_pH = hesapla_pH(m)
print(f"Hidrojen iyonu konsantrasyonu: {m}")
print(f"pH değeri: {sonuc_pH}")

def hesapla_TSS(çözünmüş_katılar, çözünmemiş_katılar):
    TSS = çözünmüş_katılar + çözünmemiş_katılar
    return TSS

çözünmüş_katılar = float(input("Çözünmüş katılar miktarını giriniz (mg/L): "))
çözünmemiş_katılar = float(input("Çözünmemiş katılar miktarını giriniz (mg/L): "))

sonuc_TSS = hesapla_TSS(çözünmüş_katılar, çözünmemiş_katılar)
print(f"Toplam Askıda Katılar: {sonuc_TSS} mg/L")

def hesapla_TDS(Elektrik_iletkenliği):
    k = 0.55
    TDS = k * Elektrik_iletkenliği
    return TDS

Elektrik_iletkenliği = float(input("Elektriksel iletkenlik değerini giriniz (µS/cm): "))

sonuc_TDS = hesapla_TDS(Elektrik_iletkenliği)
print(f"Toplam Çözünmüş Katılar (TDS): {sonuc_TDS} ppm")
```
